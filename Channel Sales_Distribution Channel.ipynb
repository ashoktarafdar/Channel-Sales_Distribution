```python
# Function to find outlier
def find_outlier_turkey(x):
    q1=x.quantile(.25)
    q3=x.quantile(.75)
    iqr=q3-q1
    floor=q1-1.5*iqr
    ceiling=q3+1.5*iqr
    outlier_indices=list(x.index[(x < floor) | (x > ceiling)])
    outlier_values=list(x[outlier_indices])
    return outlier_indices, outlier_values
```

```python
#Finding the indices of the outliers
print("Outliers for Delicatessen")
Delicatessen_indices, Delicatessen_values=find_outlier_turkey(X['Delicatessen'])
print(np.sort(Delicatessen_values))

print("Outliers for Detergents_Paper")
Detergents_Paper_indices, Detergents_Paper_values=find_outlier_turkey(X['Detergents_Paper'])
print(np.sort(Detergents_Paper_values))

print("Outliers for Frozen")
Frozen_indices, Frozen_values=find_outlier_turkey(X['Frozen'])
print(np.sort(Frozen_values))

print("Outliers for Grocery")
Grocery_indices, Grocery_values=find_outlier_turkey(X['Grocery'])
print(np.sort(Grocery_values))

print("Outliers for Milk")
Milk_indices, Milk_values=find_outlier_turkey(X['Milk'])
print(np.sort(Milk_values))

print("Outliers for Fresh")
Fresh_indices, Fresh_values=find_outlier_turkey(X['Fresh'])
print(np.sort(Fresh_values))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

def plot_corr(X,size=10):
    '''Function plots a graphical correlation matrix for each pair of coulmns in the dataframe.
    
    Input:
        df: pandas DataFrame
        size: vertical and horizontal size of the plot'''
    
    corr = X.corr()
    fig, ax = plt.subplots(figsize=(size, size))
    cax = ax.matshow(X, interpolation='nearest')
    ax.matshow(corr)
    fig.colorbar(cax)
    plt.xticks(range(len(corr.columns)), corr.columns);
    plt.yticks(range(len(corr.columns)), corr.columns);
    
plot_corr(X)
```
